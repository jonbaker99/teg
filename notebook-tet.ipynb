```python
from functools import reduce


## CREATING ROUND BESTS AND WORSTS

round_data = df.groupby(['Pl', 'Round', 'TEG'], as_index=False)[['Sc', 'Stableford', 'GrossVP', 'NetVP']].sum()
# print(round_data)

##BEST GROSS ROUND DETAILS FOR EACH PLAYER
idx_gross_min = round_data.groupby('Pl')['Sc'].idxmin()
# Use the index to select the rows with the minimum 'GrossVP' for each player
min_grossvp_details = round_data[['Pl','Round','TEG','Sc']].loc[idx_gross_min].sort_values(by='Sc', ascending=True)

print('\nBEST GROSS ROUND BY PLAYER\n')
print(min_grossvp_details)
print()

##BEST STABLEFORD DETAILS FOR EACH PLAYER
idx_stab_max = round_data.groupby('Pl')['Stableford'].idxmax()
# Use the index to select the rows with the minimum 'GrossVP' for each player
max_stab_details = round_data[['Pl','Round','TEG','Stableford']].loc[idx_stab_max].sort_values(by='Stableford', ascending=False)

print('\nBEST STABLEFORD ROUND BY PLAYER\n')
print(max_stab_details)
print()

## BEST SINGLE ROUND FOR EACH PL (NO DETAILS)
best_gross = round_data.groupby('Pl', as_index=False)['GrossVP'].min()
best_net = round_data.groupby('Pl', as_index=False)['NetVP'].min()
best_stab = round_data.groupby('Pl', as_index=False)['Stableford'].max()
best_score = round_data.groupby('Pl', as_index=False)['Sc'].min()

## COMBINING BEST ROUND DATA INTO ONE TABLE
best_dfs = [best_score,best_gross,best_net,best_stab]
pl_best_rds = reduce(lambda left, right: pd.merge(left, right, on='Pl'), best_dfs)
pl_best_rds.rename(columns={'GrossVP': 'Best gross', 'Stableford': 'Best Stableford', 'NetVP': 'Best Net', 'Sc': 'Best Score'}, inplace=True)

print('\nBEST ROUNDS\n')
print(pl_best_rds)
print()


## WORST SINGLE ROUND FOR EACH PL (NO DETAILS)
worst_gross = round_data.groupby('Pl', as_index=False)['GrossVP'].max()
worst_net = round_data.groupby('Pl', as_index=False)['NetVP'].max()
worst_stab = round_data.groupby('Pl', as_index=False)['Stableford'].min()
worst_score = round_data.groupby('Pl', as_index=False)['Sc'].max()

## COMBINING BEST ROUND DATA INTO ONE TABLE
worst_dfs = [worst_score,worst_gross,worst_net,worst_stab]
pl_worst_rds = reduce(lambda left, right: pd.merge(left, right, on='Pl'), worst_dfs)
pl_worst_rds.rename(columns={'GrossVP': 'Worst gross', 'Stableford': 'Worst Stableford', 'NetVP': 'Worst Net', 'Sc': 'Worst Score'}, inplace=True)

print('\nWORST ROUNDS\n')
print(pl_worst_rds)
print()

```

```python
## CREATING TEG BESTS AND WORSTS

## EXCLUDE TEG 2 DUE TO FEWER ROUNDS

df_ex2 = df[df['TEG'] != 'TEG 2']

teg_data = df_ex2.groupby(['Pl', 'TEG'], as_index=False)[['Sc', 'Stableford', 'GrossVP', 'NetVP']].sum()
# print(round_data)



## BEST TEG FOR EACH PL (NO DETAILS)
best_gross_teg = teg_data.groupby('Pl', as_index=False)['GrossVP'].min()
best_net_teg = teg_data.groupby('Pl', as_index=False)['NetVP'].min()
best_stab_teg = teg_data.groupby('Pl', as_index=False)['Stableford'].max()
best_score_teg = teg_data.groupby('Pl', as_index=False)['Sc'].min()

## COMBINING BEST ROUND DATA INTO ONE TABLE
best_dfs_teg = [best_score_teg,best_gross_teg,best_net_teg,best_stab_teg]
pl_best_tegs = reduce(lambda left, right: pd.merge(left, right, on='Pl'), best_dfs_teg)
pl_best_tegs.rename(columns={'GrossVP': 'Best gross', 'Stableford': 'Best Stableford', 'NetVP': 'Best Net', 'Sc': 'Best Score'}, inplace=True)

print('\nBEST TEGS\n')
print(pl_best_tegs)
print()


## WORST SINGLE ROUND FOR EACH PL (NO DETAILS)
worst_gross_teg = teg_data.groupby('Pl', as_index=False)['GrossVP'].max()
worst_net_teg = teg_data.groupby('Pl', as_index=False)['NetVP'].max()
worst_stab_teg = teg_data.groupby('Pl', as_index=False)['Stableford'].min()
worst_score_teg = teg_data.groupby('Pl', as_index=False)['Sc'].max()

## COMBINING BEST ROUND DATA INTO ONE TABLE
worst_dfs_teg = [worst_score_teg,worst_gross_teg,worst_net_teg,worst_stab_teg]
pl_worst_tegs = reduce(lambda left, right: pd.merge(left, right, on='Pl'), worst_dfs_teg)
pl_worst_tegs.rename(columns={'GrossVP': 'Worst gross', 'Stableford': 'Worst Stableford', 'NetVP': 'Worst Net', 'Sc': 'Worst Score'}, inplace=True)

print('\nWORST TEGS\n')
print(pl_worst_tegs)
print()
```
